```python
import requests #Requests is an HTTP library, written in Python, for human beings.
from bs4 import BeautifulSoup, Comment
import re
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import word_tokenize

url = "https://www.naukri.com/mnc-jobs?src=discovery_trendingWdgt_homepage_srch/index.html"
try:
  response = requests.get(url)

  #Checking if the request was successful or not
  if response.status_code==200:

    #Parsing the HTML content
    s = BeautifulSoup(response.text, "html.parser")

    #Part 1 : Discard all unnecessary tags (scripts,style,comments etc)

    for script in s(["script","style"]):
      script.extract()

    for comment in s.find_all(text=lambda text: isinstance(text, Comment)):
      comment.extract()

    #Part 2 : Extraction of all text from the HTML elements
    text = s.get_text()

    #Part 3 : Tokenization of Texts in Words
    words = word_tokenize(text)

    #Part 4 : Removal of the extraspaces and converting the text into lowercase
    cleaned_text = ''.join(words).lower()

    #Part 5 : Removal of stopwords
    nltk.download('stopwords')
    stop_words = set(stopwords.words('english'))
    words_filtered = [word for word in words if word not in stop_words]


    #Part 6 : Performing stemming and lemmatization
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    stemmed_words = [stemmer.stem(word) for word in words_filtered]
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words_filtered]

    #Part 7 : Extraction of contact information
    emailpattern = r'\b[A-Za-z0-9._%+-]+@[A-za-z09.-]+\.[A-Za|a-z]{2,7}\b'
    emails = re.findall(emailpattern, cleaned_text)


    #Part 8 : Displaying all the email addresses
    print('Extracted Email Addresses : ', emails)

    #Part 9 : Storing the cleaned and filtered data in a text file
    with open("cleaned_text.txt","w",encoding="utf-8") as file:
      file.write("Cleaned and Processed Text : \n ")
      file.write(cleaned_text)
      file.write("\n\nExtracted Email Addresses : \n")
      file.write('\n'.join(emails))


  else :
    print("Failed to retrieve the web page. Status code :", response.status_code)


except requests.exceptions.RequestException as e:
  print("An error occured during the request : ", e)

```
