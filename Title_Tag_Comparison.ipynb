```python
#Setting up the code
import requests
from bs4 import BeautifulSoup


#Function to fetch title tags from a given URL
def get_title_tags(url):
  try:
    response = requests.get(url)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
    titletag = soup.find('title')

    if titletag:
      return titletag.text.strip()

    return None

  except Exception as e:
      print(f'Error fetching title tag for {url}:{str(e)}')
  return None



#Function Block used for the comparison of the two websites
def compare_tags(hosturl,compurl):
  hosttitle = get_title_tags(hosturl)
  comptitle = get_title_tags(compurl)

  if hosttitle and comptitle:
    print(f'Title Tag for {hosturl} : {hosttitle}')
    print(f'Title Tag for {compurl} : {comptitle}')

    #Comparing them on basis of length, presence, descriptiveness and common factors etc
    length_comparison(hosttitle,comptitle)
    descriptiveness_comparison(hosttitle, comptitle)
    uniqueness_comparison(hosttitle, comptitle)
    relevance_to_content(hosttitle, comptitle)
    #branding_comparison(hosttitle, comptitle)
    #common_keywords_comparison(hosttitle, comptitle)
    #special_characters_comparison(hosttitle, comptitle)
    #character_limit_compliance(hosttitle, comptitle)
    #user_friendliness_comparison(hosttitle, comptitle)

  else:
    print('Unable to compare title tags becaise of missing data')



def length_comparison(hosttitle, comptitle):
  hostlength = len(hosttitle)
  complength = len(comptitle)

  print(f'Length Comparison: ')
  print(f'{hosturl} length :{hostlength}')
  print(f'{compurl} length : {complength}')



def descriptiveness_comparison(hosttitle, comptitle):
  if len(hosttitle) > len(comptitle):
    print(f'{hosturl} is more descriptive')
  elif len(comptitle) > len(hosttitle):
    print(f'{compurl} is more descriptive')
  else:
    print('Both tags have the same length and descriptiveness')



def uniqueness_comparison(hosttitle, comptitle):
  if hosttitle != comptitle:
    print('Title tags are unique')
  else:
    print('Title tags are not unique')



def relevance_to_content(hosttitle, comptitle):
  content_keywords = ['creative solution','edge industry knowledge','personal touch','special brand','talent acquisition','accountability','business','commitment','qualified candidate','osi staffing','recruitment challenge','risk management','staffing solution','candidate','tier candidate','recruiter','requirement','technology','comprehensive screening','admin burden','background check','job placement need','qualified professional','permanent placement','professional','streamlined onboarding','administrative burden','budget constraints']

  relevance_of_host = keyword_relevance(hosttitle, content_keywords)
  relevance_of_competitor = keyword_relevance(comptitle, content_keywords)

  print(f'Relevance to Content Comparison : ')
  print(f'{hosturl} relevance to content : {relevance_of_host}')
  print(f'{compurl} relevance to content : {relevance_of_competitor}')


  if relevance_of_host == relevance_of_competitor:
    print('Both websites have similar relevance to the content')
  elif relevance_of_host > relevance_of_competitor:
    print(f'{hosturl} is more relevant to its content')
  else:
    print(f'{compurl} is more relevant to its content')





def keyword_relevance(title, keywords):
  lowercase_title = title.lower()
  #score of relevancy
  score = sum(1 for keyword in keywords if keyword in lowercase_title)
  return score

if __name__ == "__main__":
       hosturl = 'url-example-one'
       compurl = 'url-example-two'

       compare_tags(hosturl,compurl)













```
